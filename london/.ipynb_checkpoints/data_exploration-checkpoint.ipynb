```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
import pandas as pd
df_train = pd.read_csv('/home/jbmai_sai/Documents/london/listings.csv')
df_train['price'].describe()
```

```python
print(df_train['price'].isnull().sum())
df_train['price'].fillna(df_train['price'].mode()[0], inplace=True)
print(df_train['price'].isnull().sum())
```

```python
for ind in df_train.index: 
    
    
    s = str(df_train['price'][ind])
    s1 = s.replace(',','')
    s2 = s1.replace('$','')
    df_train['price'][ind] = s2
```

```python

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
df_train.head()
```

# Univariate study (Dependent variable)

```python
df_train['price'] = df_train['price'].astype(float)
df_train['price'] = df_train['price'].astype(int)
print(df_train['price'].isnull().sum())
df_train['price'].describe()
```

```python
sns.distplot(df_train['price'])
```

# Correlation matrix (heatmap style).

```python
corrmat = df_train.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);

# in the heatmap we can observe that 
#'minumum_nights_avg_ntm'
#'minimum_minimum_nights'
#'maximum_maximum_nights'
#'minimum_maximum_nights'
#'maximum_nights_avg               --------------- are highly correlated are are 99% same (can elimate few coulmns)

#Review are also nearly correlated as it is expected (but we will include it as of now and late perform Dimentionalty reduction if required)
```

```python
#saleprice correlation matrix
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'price')['price'].index
cm = np.corrcoef(df_train[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```
